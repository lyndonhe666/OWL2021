```python
df['match_date'] = df['start_time'].map(lambda x:x[:11])
```

```python
# drop start_time column
df.drop('start_time',axis=1, inplace=True)
```

```python
df.head()
```

```python
pivoted_df = df.pivot_table(index=['esports_match_id','match_date','map_type','map_name','player_name','team_name','hero_name'], columns=['stat_name'],values=['stat_amount']).reset_index()
```

```python
pivoted_df.head()
```

```python
pivoted_df.columns
```

```python
cols_part1 = [x for x in pivoted_df.columns.get_level_values(0).tolist() if x!='stat_amount']
```

```python
pivoted_df.columns.get_level_values(1)
```

```python
cols_part2 = [x.lower().replace(' ','_') for x in pivoted_df.columns.get_level_values(1).tolist() if x!='']
```

```python
cols_part2
```

```python
chosen_features = cols_part1+cols_part2
```

```python
chosen_features
```

```python
pivoted_df.columns=chosen_features
```

```python
# filter to time played >= 120s
df1 = pivoted_df[pivoted_df['time_played']>=120]
```

```python
# filter out mistake records
df2 = df1[df1['time_alive']<=df1['time_played']]
```
